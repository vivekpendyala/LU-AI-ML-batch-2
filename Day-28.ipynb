# Calculating Recall , Precision and f1 score without sklearn

```python
recall_t = TP / (TP + FN)
precision_t = TP / (TP + FP)
f1s = 2*(recall_t*precision_t) / (recall_t + precision_t)
print('Recall :',recall_t)
print('Precision : ', precision_t)
print('f1 score :',f1s)
```
